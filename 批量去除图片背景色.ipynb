```python
import os,sys
import cv2
from PIL import Image

def re_backcolor(path,filename):
    img_path=path+'\\'+filename
    print(img_path)
    str_list=filename.split('.')
    new_path="E:\觅元素原图\免抠元素分类搜索结果\漂浮元素\\"+str_list[0]+'.png'
    print(new_path)
    img = Image.open(img_path)
    img = img.convert("RGBA")  # 转换获取信息
    pixdata = img.load()
    for y in range(img.size[1]):
        for x in range(img.size[0]):
            if pixdata[x, y][0] > 220 and pixdata[x, y][1] > 220 and pixdata[x, y][2] > 220 and pixdata[x, y][3] > 220:
                pixdata[x, y] = (255, 255, 255, 0)
    img.save(new_path)
    
    
dir_path="E:\觅元素原图\免抠元素分类搜索结果\漂浮元素"
dir_path=os.path.join(dir_path) 
img_list=os.listdir(dir_path)
i=1
for img in img_list:
    i=i+1
    re_backcolor(dir_path,img)
    print(i)
    
```
